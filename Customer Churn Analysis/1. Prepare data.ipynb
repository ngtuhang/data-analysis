## Parse log file

```python
import ast

def parse_log_line(line):
    # Convert the line to a dictionary
    data = ast.literal_eval(line)

    # Extract information
    mac = data.get('Mac', '')
    session_main_menu = data.get('SessionMainMenu', '')
    app_name = data.get('AppName', '')
    log_id = data.get('LogId', '')
    event = data.get('Event', '')
    item_id = data.get('ItemId', '')
    real_time_playing = data.get('RealTimePlaying', '')

    firmware = data.get('Firmware', '')
    default_gateway = data.get('DefaultGetway', '')
    item_name = data.get('ItemName', '')
    is_landing_page = data.get('isLandingPage', '')
    box_time = data.get('BoxTime', '')
    screen = data.get('Screen', '')
    secondary_dns = data.get('SecondaryDNS', '')
    local_type = data.get('LocalType', '')
    subnet_mask = data.get('SubnetMask', '')
    ip_wan = data.get('ip_wan', '')
    customer_id = data.get('CustomerID', '')
    contract = data.get('Contract', '')
    multicast = data.get('Multicast', '')
    primary_dns = data.get('PrimaryDNS', '')
    ip = data.get('Ip', '')
    session = data.get('Session', '')
    app_id = data.get('AppId', '')

    return [mac, session_main_menu, app_name, log_id, event, item_id, real_time_playing, 
            firmware, default_gateway, item_name, is_landing_page, box_time, screen, 
            secondary_dns, local_type, subnet_mask, ip_wan, customer_id, contract, 
            multicast, primary_dns, ip, session, app_id]

def process_log_files(log_files, output_file):
    # Define the header row
    headers = ["MAC", "SessionMainMenu", "AppName", "LogID", "Event", "ItemID", "RealTimePlaying",
               "Firmware", "DefaultGateway", "ItemName", "IsLandingPage", "BoxTime", "Screen",
               "SecondaryDNS", "LocalType", "SubnetMask", "IPWan", "CustomerID", "Contract",
               "Multicast", "PrimaryDNS", "IP", "Session", "AppID"]

    with open(output_file, 'w', encoding='utf-8') as outfile:
        # Write the header row
        outfile.write('\t'.join(headers) + '\n')

        # Process each file and write the data rows
        for file_path in log_files:
            with open(file_path, 'r', encoding='utf-8') as infile:
                for line in infile:
                    parsed_line = parse_log_line(line)
                    outfile.write('\t'.join(map(str, parsed_line)) + '\n')

# List of log files to process
log_files = [
    '/work/logt21.txt',
    '/work/logt22.txt',
    '/work/logt23.txt',
    '/work/logt24.txt',
    '/work/logt25.txt',
    '/work/logt31.txt',
    '/work/logt32.txt'
]

# Process the log files and output to a single file
process_log_files(log_files, 'parsed_log.txt')
```
