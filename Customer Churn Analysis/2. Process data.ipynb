```
df = df.drop(columns=['Firmware', 'DefaultGateway', 'IsLandingPage', 'Screen', 'SecondaryDNS', 'LocalType', 'SubnetMask', 'IPWan', 'Multicast','CustomerID','PrimaryDNS', 'IP'])
```

```
# Check for duplicates
df[df.duplicated()]
```

```
# Drop duplicates in the Dataset
df=df.drop_duplicates()
```

```
# Replace missing values in 'RealTimePlaying' with 0
df['RealTimePlaying'] = df['RealTimePlaying'].fillna(0)
```

```
# Replace missing values in 'SessionMainMenu' and 'ItemID' with 'NA'
df['SessionMainMenu'] = df['SessionMainMenu'].fillna('NA')
df['ItemID'] = df['ItemID'].fillna('NA')
```

```
# Extract datetime from 'SessionMainMenu'
df['SessionMainMenu'] = df['SessionMainMenu'].apply(lambda x: x.split(':')[1:] if isinstance(x, str) else x)
df['SessionMainMenu'] = df['SessionMainMenu'].apply(lambda x: ':'.join(x) if isinstance(x, list) else x)

# Convert 'SessionMainMenu' column to datetime
df['SessionMainMenu'] = pd.to_datetime(df['SessionMainMenu'], format='%Y:%m:%d:%H:%M:%S:%f', errors='coerce')
```

I have replaced the SessionMainMenu column value with only its date and time part and then converted it into datetime format. Now SessionMainMenu column contains only the datetime values.

## Explore user_data

```
# Read parsed_log.txt
user_info = pd.read_csv('/work/user_info.txt', sep='\t', encoding='utf-8')
```

```
# Remove 'FBOX' prefix from the MAC column
user_info['MAC'] = user_info['MAC'].str.replace('FBOX', '')
```

```
user_info.info()
```

```
# Check for duplicates
user_info[user_info.duplicated()]
```

```
# Remove duplicates MAC and keep the row with the highest '# of days'
user_info = user_info.sort_values("# of days", ascending=False).drop_duplicates('MAC')

user_info.describe()
```

## Merge data

```
# Merge user_data with log_data
HDdata = pd.merge(df, user_info, on='MAC', how='inner')
HDdata
```

```
# Save the combined data
HDdata.to_csv("HDdata.csv", index=False)
```
